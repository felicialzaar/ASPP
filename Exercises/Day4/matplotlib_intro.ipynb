```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = np.linspace(0, 4, 100)
x = np.tile(x0, (3, 1))
k = np.array([np.arange(1/2, 4*1/2, 1/2), ])
args = x*k.T*np.pi
sinx = np.sin(args)
cosx = np.cos(args)
tanx = np.tan(args)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(12, 7))
fig.suptitle("Trigonometric functions", fontsize=15)

funcs = np.array([sinx, cosx, tanx])
y_labels = ["$sin(x)$", "$cos(x)$", "$tan(x)$"]
colors = plt.cm.rainbow(np.linspace(0, 1, len(x)))
markers = ['.', '*', '+']
marker_labels = ["Dots", "Stars", "Plus signs"]
    
for i, ax in enumerate(axs.flatten()):
    ax.set_ylabel(y_labels[i], fontsize=15)
    ax.ticklabel_format(axis='y', style='sci', scilimits=(-1,2))
    
    for j, func_line in enumerate(funcs[i]):
        marker_x = [xx if xx in x0[::int(len(x0) / 20)] else float('NaN') for xx in x0]
        ax.plot(x0, func_line, color=colors[j])
        ax.plot(marker_x, func_line, marker=markers[j], color=colors[j])
        
handles = []    
legend_labels = [f"$k = ${int(kk)}$\pi$" for kk in k0]
for i, marker in enumerate(markers):
    handles.append(plt.Line2D([], [], color='k', marker=markers[i], markersize=10, label=marker_labels[i]))

for i, color in enumerate(colors):
    handles.append(plt.Line2D([], [], color=color, label=legend_labels[i]))

plt.subplots_adjust(right=0.86)
fig.legend(handles=handles, loc='center right', borderaxespad=1)
```
